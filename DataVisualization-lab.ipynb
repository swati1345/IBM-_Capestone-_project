# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
QUERY = """
SELECT ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.dropna(subset=['ConvertedComp'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
count, bin_edges = np.histogram(df)

print(count) # frequency count
print(bin_edges) # bin ranges, default = 10 bins

df.plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Developers\' Annual Salary in USD') # add a title to the histogram
plt.ylabel('Number of Respondents') # add y-label
plt.xlabel('Annual Salary') # add x-label

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

```python
df.plot(kind='box', figsize=(10,8))

plt.title('Box plot of Developers\' Age')
plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

```python
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(11, 8), color='darkblue')

plt.title('Developers\' Age and Weekly Working Hours')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import plotly.express as px
```

```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT Respondent, DatabaseWorkedWith, COUNT(DatabaseWorkedWith) as count
FROM DatabaseWorkedWith
group by Respondent
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df[(df['DatabaseWorkedWith'] == 'MySQL') & (df['count'] == 1)]
```

```python
QUERY = """
SELECT Respondent, DatabaseWorkedWith, COUNT(DatabaseWorkedWith) as count
FROM DatabaseWorkedWith
group by DatabaseWorkedWith
HAVING DatabaseWorkedWith LIKE '%SQL%'
order by count desc
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc
LIMIT 5
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

```python
fig = px.pie(df, values='count', names='DatabaseDesireNextYear', title='TOP 5 Databases Developers desire to learn next year')
fig.show()
```

```python
QUERY = """
SELECT DatabaseWorkedWith, COUNT(*) as count
FROM DatabaseWorkedWith
group by DatabaseWorkedWith
order by count desc
LIMIT 15
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df_new = df.groupby(['Age'], as_index=False).median()
df_new.set_index('Age', inplace = True)
df_new
```

```python
df_new.plot(kind='area',
             stacked=True,
             figsize=(15, 7))  # pass a tuple (x, y) size

plt.title('Median of Developers\' Weekly Work Hours and Code Review Hours in age group 30 to 35')
plt.ylabel('Hours')
plt.xlabel('Age')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df_cc = df.dropna(subset = ['ConvertedComp'], axis=0)
df_cc.reset_index(drop=True, inplace=True)
df_cc
```

```python
df_cc = df_cc.groupby(['Age'], as_index=True).median()
df_cc.plot(kind="line", figsize = (14,8))
plt.title('Median of Developers\' Annual Salary for all ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.xlabel('Age')

plt.show()
```
